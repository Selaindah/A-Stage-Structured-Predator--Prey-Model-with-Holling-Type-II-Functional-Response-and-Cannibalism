```python
import sympy as sp

# VARIABEL & PARAMETER
X, Y, Z = sp.symbols('X Y Z', real=True)

r, K = sp.symbols('r K', positive=True)
m1, n1 = sp.symbols('m1 n1', positive=True)
m2, n2 = sp.symbols('m2 n2', positive=True)
e1, e2 = sp.symbols('e1 e2', positive=True)
b, m = sp.symbols('b m', positive=True)
d1, d2 = sp.symbols('d1 d2', positive=True)

# PERSAMAAN SISTEM
f = r*X*(1 - X/K) - (m1*X*Z)/(n1 + X)
g = b*Z - (m2*Y*Z)/(n2 + Y) - (m + d1)*Y
h = e1*(m1*X*Z)/(n1 + X) + e2*(m2*Y*Z)/(n2 + Y) + m*Y - d2*Z

F = sp.Matrix([f, g, h])


J = F.jacobian([X, Y, Z])

# TITIK EQUILIBRIUM KOEKSISTENSI  E2(X,Y,Z)
Xs, Ys, Zs = sp.symbols('Xs Ys Zs', positive=True)

J_star = J.subs({X: Xs, Y: Ys, Z: Zs})


#     det(λI − J*)
λ = sp.symbols('λ')
char_poly = sp.expand((λ*sp.eye(3) - J_star).det())

P = sp.Poly(char_poly, λ)


#     λ³ + a1 λ² + a2 λ + a3 = 0
a1 = sp.simplify(P.coeffs()[1])   
a2 = sp.simplify(P.coeffs()[2])   
a3 = sp.simplify(P.coeffs()[3])   

print("\n=== KOEFISIEN POLINOMIAL KARAKTERISTIK ===")
print("a1 =", a1)
print("a2 =", a2)
print("a3 =", a3)


print("\n=== SYARAT KESTABILAN LOKAL ===")
print("1) a1 > 0")
print("2) a2 > 0")
print("3) a3 > 0")
print("4) a1*a2 > a3")
```
